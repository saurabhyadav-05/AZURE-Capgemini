```python
from google.colab import userdata
from openai import AzureOpenAI

api_key = userdata.get("OPENAI_API_KEY")

endpoint = "https://saurabh-openai1921.openai.azure.com/openai/deployments/gpt-4.1/chat/completions?api-version=2025-01-01-preview"

client = AzureOpenAI(
    api_key=api_key,
    azure_endpoint=endpoint,
    api_version="2024-02-01"
)

response = client.chat.completions.create(
    model="gpt-4.1",  # deployment name
    messages=[
        {"role": "user", "content": "Who is the president of India?"}
    ]
)

print(response.choices[0].message.content)
```

```python

# Scenario
# You’re training healthcare analysts.
# You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, and then deploy the interaction into a Gradio interface so they can try it themselves.


from google.colab import userdata
from openai import OpenAI
import gradio as gr


api_key = userdata.get("OPENAI_API_KEY")


endpoint = "https://saurabh-openai1921.openai.azure.com/"
deployment_name = "gpt-4.1"

client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}openai/deployments/{deployment_name}"
)


def healthcare_ai(question):

    completion = client.chat.completions.create(
        model=deployment_name,
        messages=[
            {
                "role": "system",
                "content": (
                    "You are a healthcare AI assistant. "
                    "Explain concepts clearly. Do NOT give diagnosis."
                ),
            },
            {
                "role": "user",
                "content": question,
            },
        ],
    )

    return completion.choices[0].message.content


interface = gr.Interface(
    fn=healthcare_ai,
    inputs=gr.Textbox(placeholder="Ask about AI in healthcare..."),
    outputs="text",
    title="🏥 Healthcare AI Assistant (Azure OpenAI)",
)

interface.launch()
```
